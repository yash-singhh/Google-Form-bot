```python
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium import webdriver
import time

# Example Google Form URL
GOOGLE_FORM_URL = "https://forms.gle/WT68aV5UnPajeoSc8"

# Path to ChromeDriver
CHROMEDRIVER_PATH = "C:\\Windows\\chromedriver.exe"

# Initialize the Chrome WebDriver
def init_webdriver():
    options = webdriver.ChromeOptions()
    # Remove '--headless' if you want to visually see the browser actions
    options.add_argument('--headless')  # Run in headless mode (optional)
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    driver = webdriver.Chrome(executable_path=CHROMEDRIVER_PATH, options=options)
    return driver

def fill_google_form():
    driver = init_webdriver()
    try:
        # Open the Google Form
        driver.get(GOOGLE_FORM_URL)
        time.sleep(3)  # Wait for the form to load

        # Fill in the "Full Name" field
        full_name_xpath = "//input[@type='text' and contains(@aria-label, 'Full Name')]"
        full_name_value = "Yash Singh"
        full_name_field = driver.find_element(By.XPATH, full_name_xpath)
        full_name_field.send_keys(full_name_value)

        # Fill in the "Contact Number" field
        contact_number_xpath = "//input[@type='text' and contains(@aria-label, 'Contact Number')]"
        contact_number_value = "811419470"
        contact_number_field = driver.find_element(By.XPATH, contact_number_xpath)
        contact_number_field.send_keys(contact_number_value)

        # Fill in the "Email ID" field
        email_id_xpath = "//input[@type='email' and contains(@aria-label, 'Email ID')]"
        email_id_value = "yash@gmail.com"
        email_id_field = driver.find_element(By.XPATH, email_id_xpath)
        email_id_field.send_keys(email_id_value)

        # Fill in the "Full Address" field
        full_address_xpath = "//textarea[contains(@aria-label, 'Full Address')]"
        full_address_value = "400074"
        full_address_field = driver.find_element(By.XPATH, full_address_xpath)
        full_address_field.send_keys(full_address_value)

        # Fill in the verification code
        verification_code_xpath = "//input[@type='text' and contains(@aria-label, 'Type this code')]"
        verification_code_value = "GNFPYC"
        verification_code_field = driver.find_element(By.XPATH, verification_code_xpath)
        verification_code_field.send_keys(verification_code_value)

        # Submit the form
        submit_button_xpath = "//div[@role='button' and contains(text(), 'Submit')]"
        submit_button = driver.find_element(By.XPATH, submit_button_xpath)
        submit_button.click()

        # Wait for the confirmation page to load
        time.sleep(3)

        # Take a screenshot of the confirmation page
        confirmation_screenshot_path = "confirmation_page.png"
        driver.save_screenshot(confirmation_screenshot_path)
        print(f"Screenshot of the confirmation page saved as {confirmation_screenshot_path}")

        print("Google Form submitted successfully!")

    except Exception as e:
        print(f"An error occurred: {e}")

    finally:
        driver.quit()

if __name__ == "__main__":
    fill_google_form()
```
